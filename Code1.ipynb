```python
import pandas as pd
```

```python
data = pd.read_csv('linear_vectorfield_data_x0.txt',header=None)
```

```python
data['k']=data[0].str.split().str[0]
data['x0']=data[0].str.split().str[1]
```

```python
data.drop(columns=0,inplace=True)
```

```python
data1 = pd.read_csv('linear_vectorfield_data_x1.txt',header=None)
data1['k']=data1[0].str.split().str[0]
data1['x1']=data1[0].str.split().str[1]
data1.drop(columns=0,inplace=True)
```

```python
import numpy as np
x=data['x0']
y=data1['x1']
```

```python
x=np.array(x,dtype=float)
y=np.array(y,dtype=float)
```

```python
A = np.vstack([x, np.ones(len(x))]).T
```

# Matrix A

```python
A
```

```python
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
```

```python
m,c
```

```python
import matplotlib.pyplot as plt
_ = plt.plot(x, y, 'o', label='Original data', markersize=10)

plt.plot(x, m*x + c, 'r', label='Fitted line')


plt.show()
```

```python
x1_estimated = m*x 
```

# Mean Squared Error

```python
k=np.array([0])
for i in range(1000):
    k=k+np.square(y[i]-x1_estimated[i])
```

```python
Error=k/1000
Error[0]
```

```python
from numpy import exp, loadtxt, pi, sqrt
import matplotlib.pyplot as plt
import numpy as np
import scipy  as scipy
```

```python
data=loadtxt('/Users/narminmammadova/Desktop/TUMdersler/lab/exercise5/linear_function_data.txt')
x=data[:,0]
y=data[:,1]
```

```python
# approximating data A with linear function
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
plt.plot(x, y, 'o', label='Original data', markersize=10)
plt.plot(x, m*x + c, 'r', label='Fitted line')
plt.legend()
```
